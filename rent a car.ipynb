```python
# python assignment
# car renting system

print("\n Hip Hop Rentals")

# void function with one parameter
def welcome_customer(customer_name):
    print(f"Welcome to Hip Hop Rentals {customer_name} ! thank you for choosing our serivce.")

#fruitful function without prameter
import random
def generate_reservation_number():
    return f"RES{random.randint(1000, 9999)}"

# fruitful function with two parameters,nested conditionals and while loop 
def rent_car(customer_name, rental_duration):
    if rental_duration <= 0:
        print("Invalid Rental Duration! please provide a valid rental duration.")
        return None

    available_cars = {"Compact": 10, "Sedan": 8, "SUV": 5}
    selected_car_type = input("Enter the car type (Compact/Sedan/SUV): ").capitalize()

    while selected_car_type not in available_cars or available_cars[selected_car_type] == 0:
        print("Sorry! The selected_car_type is not available or out of stock.")
        selected_car_type = input("Enter a different car type (Compact/Sedan/SUV): ").capitalize()

    rental_cost = 0
    if rental_duration <=3:
        rental_cost = 30 * rental_duration
    elif rental_duration <=7:
        rental_cost = 25 * rental_duration
    else:
        rental_cost = 20 * rental_duration

    available_cars[selected_car_type] -= 1
    reservation_number = generate_reservation_number()

    return {"customer_name": customer_name, "car_type": selected_car_type, "rental duration": rental_duration,
    "rental_cost": rental_cost, "reservation_number": reservation_number}

# anonymous function

calculate_discount = lambda total_cost, discount_rate: total_cost * discount_rate / 100

# output

customer_name = input("Enter your name: ")
welcome_customer(customer_name)

rental_duration = int(input("Enter the rental duration (in days): "))
rental_details = rent_car(customer_name, rental_duration)

if rental_details:
    print("\n Rental Details")
    print(rental_details)
    print("Total Rental Cost: $", rental_details["rental_cost"])

    discount_rate = 10
    discounted_cost = calculate_discount(rental_details["rental_cost"], discount_rate)
    print(f"Discounted Cost ({discount_rate} %): $: ", discounted_cost)

    
    
# String methods
print("\n Strings Methods")
Confirmation_message = "Your car reservation is confirmed.Enjoy your ride!"
print(Confirmation_message)
uppercase = Confirmation_message.upper()
print("Uppercase of this string is", uppercase)
lowercase = Confirmation_message.lower()
print("Uppercase of this string is", lowercase)
sub_string_message = Confirmation_message[5:18]
print("Sub string of this string is", sub_string_message)
message_length = len(Confirmation_message)
print("message_length of this string is", message_length)
    
print("\n python collections")

# using collections
print("\n List")
# List

cars_available = ["Compact", "Sedan", "SUV"]
print(cars_available)
cars_available.append("Truck")
print("It will append one more vehicle in the end of the list", cars_available)
cars_available.extend(["Convertable", "Van"])
print("it will extend the list", cars_available)
cars_available.remove("Compact")
print("it will remove the first occurance of the value", cars_available)
cars_available.pop()
print("it will remove the last value from the list", cars_available)

# Dictionary
print("\n Dictionary")
cars_available = {"Compact": 10, "Sedan": 8, "SUV": 5}
print("keys of this dictionary is ",cars_available.keys())
print("values of this dictionary is ", cars_available.values())
print("it will remove the targeted dictionary ", cars_available.pop("SUV"))
print("it will remove the last dictionary",cars_available.popitem())
```
